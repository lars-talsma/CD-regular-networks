```python
import numpy as np
import matplotlib.pyplot as plt

col = plt.cm.viridis([0, 0.45 , 0.9])    

plt.style.use('figure.mplstyle')

mm = 1/25.4
outward = 4
color_text = (0, 0, 0, 0.8)
fs = 8
tudelft_sea_green = '#6ec8bb'

fig, ax = plt.subplots(1, 1, constrained_layout=True, figsize=[86*mm, 86*mm/3])

ax.spines[['bottom', 'right', 'top', 'left']].set_visible(False)

ax.set_yticks([])
ax.set_xticks([])

Delta_x = 0.4

x_0 = 0.05
y_0 = 0
delta_x = 0.08
delta_y = delta_x

ms=12
lw=3
a=0.4
va=1

# Chain
ax.plot([x_0+0.5*delta_x, x_0+1.5*delta_x, x_0+2.5*delta_x], [y_0, y_0, y_0], '.', markersize=ms, color=tudelft_sea_green)
ax.plot([x_0+0.5*delta_x, x_0+1.5*delta_x, x_0+2.5*delta_x], [y_0, y_0, y_0], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
ax.text(x_0+1.5*delta_x, y_0-va*delta_y, '$d=2$', fontsize=fs, ha='center', color=color_text)


# Honeycomb
for i in range(6):
    # top honeycomb
    ax.plot(x_0+(1+1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos(i*np.pi/3), y_0+delta_y*(np.sin(i*np.pi/3)+np.sin(np.pi/3)), '.', markersize=ms, color=tudelft_sea_green)
    # bottom honeycomb
    ax.plot(x_0+(1+1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos(i*np.pi/3), y_0+delta_y*(np.sin(i*np.pi/3)-np.sin(np.pi/3)), '.', markersize=ms, color=tudelft_sea_green)
    # left honeycomb
    ax.plot(x_0+(1-1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos(i*np.pi/3), y_0+delta_y*(np.sin(i*np.pi/3)), '.', markersize=ms, color=tudelft_sea_green)

for i in range(5):
    ax.plot([x_0+(1+1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos(i*np.pi/3), x_0+(1+1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos((i-1)*np.pi/3)], [y_0+delta_y*(np.sin(i*np.pi/3)+np.sin(np.pi/3)), y_0+delta_y*(np.sin((i-1)*np.pi/3)+np.sin(np.pi/3))], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
    ax.plot([x_0+(1+1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos((i+3)*np.pi/3), x_0+(1+1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos((i+4)*np.pi/3)], [y_0+delta_y*(np.sin((i+3)*np.pi/3)-np.sin(np.pi/3)), y_0+delta_y*(np.sin((i+4)*np.pi/3)-np.sin(np.pi/3))], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
    ax.plot([x_0+(1-1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos((i+1)*np.pi/3), x_0+(1-1.5*np.cos(np.pi/3))*delta_x+Delta_x+delta_x*np.cos((i+2)*np.pi/3)], [y_0+delta_y*(np.sin((i+1)*np.pi/3)), y_0+delta_y*(np.sin((i+2)*np.pi/3))], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)

    
ax.text(x_0+delta_x+Delta_x, y_0-(va+1+np.sin(np.pi/3))*delta_y, '$d=3$', fontsize=fs, ha='center', color=color_text)


# Square
for i in range(3):
    ax.plot([x_0-delta_x/2+2*Delta_x+0*delta_x, x_0-delta_x/2+2*Delta_x+2*delta_x], [y_0-delta_y+i*delta_y, y_0-delta_y+i*delta_y], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
    ax.plot([x_0-delta_x/2+2*Delta_x+i*delta_x, x_0-delta_x/2+2*Delta_x+i*delta_x], [y_0-delta_y+0*delta_y, y_0-delta_y+2*delta_y], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
    for j in range(3):
        ax.plot([x_0-delta_x/2+2*Delta_x+i*delta_x], [y_0-delta_y+j*delta_y], '.', markersize=ms, color=tudelft_sea_green)

ax.text(x_0+0.5*delta_x+2*Delta_x, y_0-(1+va)*delta_y, '$d=4$', fontsize=fs, ha='center', color=color_text)


# Hexagonal
y_corr = -np.sin(np.pi/3)*delta_y/2
ax.plot(x_0+3*Delta_x, y_0+y_corr, '.', markersize=ms, color=tudelft_sea_green)
ax.plot(x_0+3*Delta_x, y_0+y_corr, '.', markersize=ms, color=tudelft_sea_green)
ax.plot(x_0+3*Delta_x+(1+np.cos(np.pi/3))*delta_x, y_0-np.sin(np.pi/3)*delta_y+y_corr, '.', markersize=ms, color=tudelft_sea_green)
ax.plot(x_0+3*Delta_x-(1+np.cos(np.pi/3))*delta_x, y_0-np.sin(np.pi/3)*delta_y+y_corr, '.', markersize=ms, color=tudelft_sea_green)

ax.plot([x_0+3*Delta_x-(1+np.cos(np.pi/3))*delta_x, x_0+3*Delta_x-(np.cos(np.pi/3))*delta_x], [y_0-np.sin(np.pi/3)*delta_y+y_corr, y_0-np.sin(np.pi/3)*delta_y+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
ax.plot([x_0+3*Delta_x+(1+np.cos(np.pi/3))*delta_x, x_0+3*Delta_x+(np.cos(np.pi/3))*delta_x], [y_0-np.sin(np.pi/3)*delta_y+y_corr, y_0-np.sin(np.pi/3)*delta_y+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)

ax.plot([x_0+3*Delta_x-(1+np.cos(np.pi/3))*delta_x, x_0+3*Delta_x-(1)*delta_x], [y_0-np.sin(np.pi/3)*delta_y+y_corr, y_0+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
ax.plot([x_0+3*Delta_x+(1+np.cos(np.pi/3))*delta_x, x_0+3*Delta_x+(1)*delta_x], [y_0-np.sin(np.pi/3)*delta_y+y_corr, y_0+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)

ax.plot(x_0+3*Delta_x, y_0+2*np.sin(np.pi/3)*delta_y+y_corr, '.', markersize=ms, color=tudelft_sea_green)

ax.plot([x_0+3*Delta_x, x_0+3*Delta_x-(np.cos(np.pi/3))*delta_x], [y_0+2*np.sin(np.pi/3)*delta_y+y_corr, y_0+1*np.sin(np.pi/3)*delta_y+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
ax.plot([x_0+3*Delta_x, x_0+3*Delta_x+(np.cos(np.pi/3))*delta_x], [y_0+2*np.sin(np.pi/3)*delta_y+y_corr, y_0+1*np.sin(np.pi/3)*delta_y+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)

for i in range(6):
    ax.plot(x_0+3*Delta_x+delta_x*np.cos(i*np.pi/3), y_0+delta_y*np.sin(i*np.pi/3)+y_corr, '.', markersize=ms, color=tudelft_sea_green)

    ax.plot([x_0+3*Delta_x+delta_x*np.cos(i*np.pi/3), x_0+3*Delta_x+delta_x*np.cos((i+1)*np.pi/3)], [y_0+delta_y*np.sin(i*np.pi/3)+y_corr, y_0+delta_y*np.sin((i+1)*np.pi/3)+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)
    ax.plot([x_0+3*Delta_x+delta_x*np.cos(i*np.pi/3), x_0+3*Delta_x], [y_0+delta_y*np.sin(i*np.pi/3)+y_corr, y_0+y_corr], '-', linewidth=lw, color=tudelft_sea_green, alpha=a)

ax.text(x_0+3*+Delta_x, y_0-(va+1.5*np.sin(np.pi/3))*delta_y, '$d=6$', fontsize=fs, ha='center', color=color_text)

ax.set_aspect(1)
ax.set_ylim([-0.21, 0.21])

# fig.savefig('figures/topology.pdf', bbox_inches = 'tight', pad_inches=0*mm)
```
